```python
import pandas as pd
```

```python
df = pd.read_html('../RICERCADOCENTI.html')[0]
```

```python
df.columns
```

```python
fascia_simplified = df["Fascia"].astype(str).str.split().str[0]

stacked_counts = (
  df.assign(Fascia_grouped=fascia_simplified.replace({"Straordinario": "Ordinario"}))
    .groupby(["Fascia_grouped", "Genere"])
    .size()
    .unstack(fill_value=0)
    .reindex(["Ordinario", "Associato", "Ricercatore"], fill_value=0)
)

ax = stacked_counts.plot(kind="bar", stacked=True, figsize=(8, 5))
ax.set_xlabel("Fascia")
ax.set_ylabel("Count")
ax.set_title("Distribuzione per Fascia e Genere")
```

```python
from html import escape
from IPython.display import HTML

_STYLE = """<style>
.vt { border-collapse: collapse !important; margin: .5em 0 !important; background: white !important; text-align: left !important; }
.vt th { border: 1px solid #ccc !important; padding: 3px 8px !important; font-weight: bold !important; white-space: nowrap !important; background: white !important; text-align: left !important; }
.vt td { border: 1px solid #ccc !important; padding: 3px 8px !important; background: white !important; text-align: left !important; }
.vs { max-width: 300px !important; overflow: hidden !important; text-overflow: ellipsis !important; white-space: nowrap !important; }
</style>"""

def _render(v, prefix=''):
  if isinstance(v, dict):
    return f'<details><summary>{prefix}{{…}}</summary>{_visit(v)}</details>'
  if isinstance(v, list):
    return f'<details{" open" if len(v) == 1 else ""}><summary>{prefix}[…]</summary>{_visit(v)}</details>'
  return f'<div class="vs">{prefix}{escape(str(v))}</div>'

def _visit(data):
  if isinstance(data, dict):
    return f'<table class="vt">{"".join(f'<tr><th>{escape(str(k))}</th><td>{_render(v)}</td></tr>' for k, v in data.items())}</table>'
  if isinstance(data, list):
    return f'<div style="padding-left:1em">{"".join(_render(v, prefix=f'[{i}] ') for i, v in enumerate(data))}</div>'
  return _render(data)

def visit(data):
  return HTML(_STYLE + _visit(data))
```
